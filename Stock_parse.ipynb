```
def find(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]
```

```
from tqdm import tqdm

searches = []

for i in r.text.split('\n'):
    if 'href="/search/' in i:
      indc = find(i, '"')
      c_slice = i[indc[0]+1:indc[1]]
      if '/' in c_slice:
          searches.append(c_slice)

print(len(searches))
searches = set(searches)
print(len(searches))
```

```
import time
```

```
current_images = []

for elem in tqdm(searches):
    url = 'https://stocksnap.io' + elem
    r=requests.get(url)

    for i in r.text.split('\n'):
        if 'photo-grid-preview' in i:
            indc = find(i, '"')
            c_slice = i[indc[0]+1:indc[1]]
            current_images.append(c_slice)

    time.sleep(0.5)
```

```
print(len(current_images))
current_images = set(current_images)
print(len(current_images))
```
